```python
#This algorithm is going to put all of the seasons links in one text so we will be able to go through them
#and get each season all of the game links and then
#we will be able to go through every game link and take data

#1995-2013 is &se=4 -> &se=22 therefore first loop will begin from 9 to 23
season = 9
sitesofseasons=[]
website = "https://www.premierleague.com/results?"
parameter_1 = "co=1&"
parameter_2 = "se="
parameter_3 = "&cl=-1"
for season in range(4,23):
    sitesofseasons.append(website+parameter_1+parameter_2+str(season)+parameter_3)
#2014/15 = 27, 2015/16=42, 2016/17=54, 2017/18=79
sitesofseasons.append(website+parameter_1+parameter_2+str(27)+parameter_3)
sitesofseasons.append(website+parameter_1+parameter_2+str(42)+parameter_3)
sitesofseasons.append(website+parameter_1+parameter_2+str(54)+parameter_3)
sitesofseasons.append(website+parameter_1+parameter_2+str(79)+parameter_3)

with open('Seasons-links.txt', 'w') as filehandle:#opening a file and putting all of the season links in them
    filehandle.writelines("%s\n" % site for site in sitesofseasons)
```

```python
#creating a dictionary so we will identify which link is the season we are speaking about.
seasons_dict = {}
first_year=95
first_year_str = ''
second_year=96
second_year_str = ''
for i in range(19):
    first_year = first_year%100
    #print(first_year)
    second_year = second_year%100
    #print(second_year)
    if(first_year<10):
        #print("First_year is " + str(first_year))
        first_year_str = '0' + str(first_year)
        #print(first_year_str)
    else:
        first_year_str = str(first_year)
    if(second_year<10):
        #print("Second_year is " + str(second_year))
        second_year_str = '0' + str(second_year)
        #print(second_year_str)
    else:
        second_year_str = str(second_year)
    seasons_dict[sitesofseasons[i]]= first_year_str + '' + second_year_str
    first_year+=1
    second_year+=1

seasons_dict[sitesofseasons[19]]="1415"
seasons_dict[sitesofseasons[20]]="1516"
seasons_dict[sitesofseasons[21]]="1617"
seasons_dict[sitesofseasons[22]]="1718"
```

```python
def press_accept_cookies(driver):
    l=driver.find_element_by_class_name("_2hTJ5th4dIYlveipSEMYHH ")
    #we detect the button of accept cookies
    l.click()
    #we press on accept cookies
```

```python
#using this algorithm to get all of the games links in the last 20 years, we are getting also those who 
#had lags in the middle of information taking and we are retaking it.

import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


import time
PATH = "C:\Program Files (x86)\chromedriver.exe"
#PATH of webdriver in the pc
driver = webdriver.Chrome(PATH)
#creating webdriver object named driver

#LOOOPPPPPPPPP
did_you_press_accept_cookies=0
for current_season in seasons_dict:
    text_file_name = 'Matches-' + seasons_dict[current_season] + '-links.txt'
    driver.get(current_season) 
    #opening the current season link from the sitesofseasons list
    driver.maximize_window()
    #maximizing the window
    
    #we wait for the accept cookies offer to pop up
    if(did_you_press_accept_cookies==0):
        driver.implicitly_wait(1)
        press_accept_cookies(driver)
        did_you_press_accept_cookies+=1
        
    driver.implicitly_wait(5)

    element = driver.find_element_by_tag_name("body")
    #we detect the body
    for number_of_pages in range(50):
        element.send_keys(Keys.PAGE_DOWN)
        time.sleep(1)
        #we go down in 50 seconds so we will have all of the season fixture game results loaded up

        fixtures_list = driver.find_elements_by_class_name("matchList")
        #we take a match_list object which contains all of the fixtures
        links = []
    for current_fixture in fixtures_list:
        current_match = current_fixture.find_elements_by_class_name("matchFixtureContainer")
        for match in current_match:
            matchId = match.get_attribute("data-comp-match-item")
            links.append("https://www.premierleague.com/match/"+ matchId)
        #what we did here was to get in every fixture in the fixtures list
        #we took every match from every fixture in the fixtures list and appended it the links list
        #after that we needed to make sure there was no duplicate links.
    links = list( dict.fromkeys(links) )#removes duplicates from the links list
    with open(text_file_name, 'w') as filehandle:#opening a file and putting all of the matches links in them
        filehandle.writelines("%s\n" % link for link in links)#writing each link with \n into a text file so we could run on
    #start run on it and fetch every game specific details we need
    time.sleep(5)
```
