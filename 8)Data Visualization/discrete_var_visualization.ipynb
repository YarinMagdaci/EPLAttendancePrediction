```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('bigData_cleaned2.csv', header=0, sep=',') 
```

```python
#Here we can see the ratio between number of games which are derbies in orange,
#And the not derbies games in blue, kind of like pacman shape
data["derby"].value_counts().plot(kind='pie')
```

```python
#Same, we want to see how many games are derbies and how many are not
#using bars this time, Y axis is the number of games eventually.
mycolors = ['#008000', '#FF0000']
data["derby"].value_counts().plot(kind='bar', color = mycolors)
plt.ylabel('Games')
plt.xlabel('Derby')
plt.title('Derbies against not derbies')
```

```python
#Here we see how many games we have for each team hoster rating.
data["team_1_rating"].value_counts().plot(kind='barh')
plt.title('Number of teams by rating')
```
