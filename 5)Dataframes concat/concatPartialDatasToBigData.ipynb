```python
#here we just created a system to go over our csv files we have saved with partial data files
#our main goal in this program here is to merge all the data to one big data
#we use pandas, use some manipulations on strings so we could work faster
#eventually we create here a list of dataframes and after this cell we will concat them easily by rows [axis=0]
#then we will print the bigData and we will see if anything needs to be done, if any thing will not be helpful we will remove it.

import pandas as pd
file_name_part_one  = 'this_df_'
file_name_part_three = '.csv'
dataframes = list()
for digit in range(12):
    file_name = file_name_part_one + str(digit) + file_name_part_three
    dataframes.append(load_dataFrame(file_name))
```

```python
def load_dataFrame(name):
    return pd.read_csv(name)
```

```python
big_Dataframe = pd.concat(dataframes, axis=0)#concat all the dataframes we had to one big dataframe by rows [axis=0]
```
